# 02 — Spatial Panel Construction

Build the segment panel and run the four spatial joins.

```python
%matplotlib inline
import pandas as pd
```

```python
from phillyparking.spatial.segments import build_segments, save_segments
from phillyparking.spatial.zones import assign_zones, zone_summary
from phillyparking.spatial.join import join_acs, join_lodes, join_transit, join_pois
from phillyparking._config import data_dir
```

## Build segments

```python
segments = build_segments()
print('segments:', segments.shape)
segments.head()
```

## Spatial joins

```python
segments = join_acs(segments)
segments = join_lodes(segments)
segments = join_transit(segments)
segments = join_pois(segments)
print('after joins:', segments.shape)
segments.head()
```

## Zone assignment

```python
segments = assign_zones(segments)
summary = zone_summary(segments)
summary
```

## Inline segment map

```python
try:
    from phillyparking.viz.maps import segment_map
    m = segment_map(segments.head(200))
    m
except ImportError:
    print('Install folium to render the segment map.')
```

## Persist to processed parquet

```python
out = data_dir() / 'processed' / 'segments.parquet'
out.parent.mkdir(parents=True, exist_ok=True)
save_segments(segments, out)
print('wrote', out)
```
