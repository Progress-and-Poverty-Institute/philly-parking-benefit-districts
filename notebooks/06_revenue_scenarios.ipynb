# 06 — Revenue Scenarios

Run all elasticity scenarios, build the Laffer curve, and split revenue into the PBD share.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from phillyparking.scenarios.runner import run_all_scenarios
from phillyparking.revenue.forecast import revenue_curve, revenue_under_scenarios
from phillyparking.revenue.allocation import allocate
from phillyparking.viz.charts import laffer_curve
results = run_all_scenarios()
list(results.keys())
```

## Total revenue per scenario

```python
rev = {k: float(v.get('revenue', np.nan)) for k, v in results.items()}
rev_s = pd.Series(rev, name='revenue')
fig, ax = plt.subplots(figsize=(6, 4))
rev_s.plot.bar(ax=ax)
ax.set_ylabel('annual revenue ($)')
ax.set_title('Revenue by scenario')
plt.tight_layout(); plt.show()
rev_s
```

## Laffer curve for CCC

```python
try:
    curve = revenue_curve(zone='CCC', multipliers=np.linspace(0.4, 2.5, 25))
    fig = laffer_curve(curve)
    plt.show()
except Exception as e:
    print('Laffer curve unavailable:', e)
```

## Revenue under three elasticity scenarios with 90% bootstrap CIs

```python
try:
    band = revenue_under_scenarios(n_boot=200, ci=0.90)
    band
except Exception as e:
    print('bootstrap unavailable:', e)
```

## PBD allocation

```python
total = float(rev_s.sum())
alloc = allocate(total, pbd_share=0.30)
alloc
```
