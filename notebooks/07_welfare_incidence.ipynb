# 07 — Welfare & Incidence

Cruising deadweight loss and a placeholder distributional incidence calculation.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from phillyparking.scenarios.runner import run_all_scenarios
from phillyparking.welfare.cruising_dwl import annual_cruising_dwl
from phillyparking.welfare.incidence import gini, incidence_summary
results = run_all_scenarios()
```

## Annual cruising DWL per scenario

```python
dwl = {}
for name, res in results.items():
    try:
        dwl[name] = float(annual_cruising_dwl(res['segments'], res['panel']))
    except Exception as e:
        dwl[name] = float('nan')
        print(name, 'failed:', e)
pd.Series(dwl, name='cruising_dwl_usd')
```

## Synthetic decile-uniform incidence

```python
rng = np.random.default_rng(0)
n = 5000
incidence_df = pd.DataFrame({
    'household_id': range(n),
    'income_decile': rng.integers(1, 11, size=n),
    'parking_spend_usd': rng.gamma(shape=2.0, scale=120.0, size=n),
})
summary = incidence_summary(incidence_df)
summary
```

## Gini on synthetic distribution

```python
g = gini(incidence_df['parking_spend_usd'].values)
print(f'Gini = {g:.3f}')
```
