# 09 — Scenario Comparison

Final roll-up: compare elasticity scenarios on revenue, cruising DWL, mean price, and PBD allocation.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from phillyparking.scenarios.runner import run_all_scenarios
from phillyparking.scenarios.compare import comparison_table
results = run_all_scenarios()
table = comparison_table(results)
table
```

## Small multiples across scenarios

```python
metrics = [c for c in ['revenue', 'cruising_dwl', 'mean_price', 'pbd_allocation']
           if c in table.columns]
fig, axes = plt.subplots(1, len(metrics), figsize=(4 * len(metrics), 4))
if len(metrics) == 1:
    axes = [axes]
for ax, metric in zip(axes, metrics):
    table[metric].plot.bar(ax=ax)
    ax.set_title(metric)
plt.tight_layout(); plt.show()
```
