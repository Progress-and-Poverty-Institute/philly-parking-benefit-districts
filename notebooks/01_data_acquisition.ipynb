# 01 — Data Acquisition

Walk through every public data source the model consumes. Each loader is wrapped in `try/except` so the notebook still runs offline (synthetic fallbacks kick in inside the `phillyparking.io` modules).


```python
%matplotlib inline
import pandas as pd
```

```python
from phillyparking.io import opendataphilly, census, lodes, osm_loader, gtfs_loader, ppa_stub
```

## Meters (OpenDataPhilly)

```python
try:
    meters = opendataphilly.load_meters(refresh=False)
    print('meters:', meters.shape)
    display(meters.head())
except Exception as e:
    print('meters fallback:', e)
```

## Streets (OpenDataPhilly)

```python
try:
    streets = opendataphilly.load_streets(refresh=False)
    print('streets:', streets.shape)
    display(streets.head())
except Exception as e:
    print('streets fallback:', e)
```

## ACS demographics (Census)

```python
try:
    acs = census.load_acs(refresh=False)
    print('acs:', acs.shape)
    display(acs.head())
except Exception as e:
    print('acs fallback:', e)
```

## LODES origin–destination jobs

```python
try:
    lodes_df = lodes.load_lodes(refresh=False)
    print('lodes:', lodes_df.shape)
    display(lodes_df.head())
except Exception as e:
    print('lodes fallback:', e)
```

## OSM points-of-interest

```python
try:
    pois = osm_loader.load_pois(refresh=False)
    print('pois:', pois.shape)
    display(pois.head())
except Exception as e:
    print('pois fallback:', e)
```

## GTFS transit stops

```python
try:
    stops = gtfs_loader.load_stops(refresh=False)
    print('stops:', stops.shape)
    display(stops.head())
except Exception as e:
    print('stops fallback:', e)
```

## PPA transactions (stub)

```python
try:
    txns = ppa_stub.load_transactions()
    print('txns:', txns.shape)
    display(txns.head())
except Exception as e:
    print('ppa fallback:', e)
```
